```python
import scipy
import scipy.stats
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Files/bicicletas-compartidas.csv')
```

Categoricas

```python
df.columns
```

```python
y = df['cuartil-ano'].apply(lambda x: 'cuartil-' + str(int(x)))
y 
```

```python
fig, ax = plt.subplots() #los parametros de subplots son fig y ax
ax.bar(y.value_counts().index, y.value_counts())
ax.set_xlabel('Cuartiles del ano')
ax.set_ylabel('Frecuencia')
ax.patches[3].set_facecolor('green')#solo si queremos marcar o agregar color y patches es el numero que queremos marcar
#python ordena los datos de mayor a menor, por eso no estan en orden
```

Numericas

```python
y = df['viento']
```

```python
fig, ax = plt.subplots()
ax.hist(y, bins = 30)#hist de histograma
ax.set_xlabel('viento')# X label
ax.set_ylabel('frequencia') # Y label
plt.axvline(np.mean(y), c = 'r', linestyle = '--', label = 'promedio')#recta que representa la media
#  ax v(vertical)line, np.mean(funcion de a media), c(de color)
plt.axvline(np.mean(y) + np.std(y), c = 'k', linestyle = '--', label = '+ 1-desviacion')
plt.axvline(np.mean(y) - np.std(y), c = 'g', linestyle = '--', label = '- 1-desviacion')
ax.legend()
```

Boxplot y scatterplot

```python
fig, ax = plt.subplots()
sns.boxplot(x = 'cuartil-ano', y = 'bicis-compartidas', data = df)
```

```python
fig, ax = plt.subplots()
#para scatter plt se necesitan 2 variables continuas
ax.scatter(df['viento'],df['bicis-compartidas'], alpha = 0.03)
ax.set_title('distribucion conjunta de viento y bicicletas compartidas')
ax.set_xlabel('viento')
ax.set_ylabel('bicis-compartidas')
```
